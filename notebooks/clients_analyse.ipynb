```python
from azure.storage.blob import BlobServiceClient
import pandas as pd
from io import StringIO
```

```python
connection_string = ""

# Nom du container
container_name = "76byj86oc9kf"
blob_service_client = BlobServiceClient.from_connection_string(connection_string)

# Récupère le container
container_client = blob_service_client.get_container_client(container_name)

clients_file_name = "clients.csv"
```

```python
blob_client = container_client.get_blob_client(clients_file_name)
blob_data_client = blob_client.download_blob().readall()
df_clients = pd.read_csv(StringIO(blob_data_client.decode('utf-8')), sep=';')

print(df_clients.head())
print(df_clients.count())
```

```python
df_clients_without_duplicate = df_clients[~df_clients.duplicated(subset='id', keep='first')].copy()

# Convertir les colonnes en string
df_clients_without_duplicate['id'] = df_clients_without_duplicate['id'].astype(str)
df_clients_without_duplicate['account_id'] = df_clients_without_duplicate['account_id'].astype(str)

# Renommer les colonnes
df_clients_without_duplicate = df_clients_without_duplicate.rename(columns={
    'id': 'clientId',
    'account_id': 'accountId'
})


```
